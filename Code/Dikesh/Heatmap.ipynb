```python
#Get data from csv file

import pandas as pd

df = pd.read_csv("../../Data/Nepal_RoadAccidents_Cleaned.csv")

df.head(3)
```

Heatmap of major accident places in nepal

```python

import folium
from folium.plugins import HeatMap


map_center = [df["Latitude"].mean(), df["Longitude"].mean()]  #Finds the average location of all the accidents

m = folium.Map(location=map_center, zoom_start=7) #Creates a folium map centered around the average location 

heat_data = df[["Latitude", "Longitude"]].values.tolist() #Converts the latitude and longitude columns into a list of lists

HeatMap(heat_data).add_to(m) #Adds the heatmap layer to the folium map

m
```

Temporal Distribution of Accidents

```python
time_counts = df["Time_of_Day"].value_counts()
print(time_counts)
```

```python
import matplotlib.pyplot as plt

time_counts.plot(kind="bar")

plt.title("Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")

plt.show()
```

Accidents by Road type

```python
road_counts = df["Road_Type"].value_counts()
print(road_counts)

road_counts.plot(kind="bar")

plt.title("Accidents by Road Type")
plt.xlabel("Road Type")
plt.ylabel("Number of Accidents")

plt.show()
```

Analyze Accidents by Weather Condition

```python
weather_counts = df["Weather_Condition"].value_counts()
print(weather_counts)

weather_counts.plot(kind="bar")

plt.title("Accidents by Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Number of Accidents")

plt.show()
```
